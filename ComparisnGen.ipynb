```python
def createRatingDataframe(sourceDf,playerDict):
    defObj = {'clubName':'','player_id':0, 'player':'', 'pass':0,'successfulPass':0, 'shot':0, 
              'cross':0, 'through_ball':0,'dribble':0, 'assist':0,'blockpass':0,'foulConceded':0,
              'goal':0, 'ttime':0,'yellow_card':0,'red_card':0,'tackles':0,'tackleSuccess':0,'blocks':0,'saves':0}
    data = sourceDf.groupby('game_id')
    for key, item in data:
        matchObj = data.get_group(key)
        matchObj = matchObj.dropna(subset=['player_id'])
        ctr=0
        for idx,rowObj in matchObj.iterrows():
            playerId = (rowObj['player_id'])
            if playerId == '':
                continue;
            else:
                playerId = int(playerId)
            if playerId in playerDict:
                playerObj = playerDict[(playerId)]
            else:
                playerObj = defObj.copy()

            if playerObj['player_id'] == 0:
                playerObj['player_id'] = playerId
            if playerObj['player'] == '':
                playerObj['player'] = rowObj['player']
            playerObj['clubName'] = rowObj['team']
            event_id = rowObj['event_type_id']
            if event_id == 1:
                playerObj['pass'] += 1
                if(rowObj['outcome'])==1:
                    playerObj['successfulPass'] += 1
                if rowObj['cross'] == 1:
                    playerObj['cross']+=1
                if rowObj['assist'] == 1:
                    playerObj['assist']+=1
                if rowObj['through_ball']==1:
                    playerObj['through_ball']+=1
            elif event_id in[13,16,15,14]:
                playerObj['shot']+=1
                if event_id == 16:
                    playerObj['goal']+=1
            elif event_id == 3 and rowObj['outcome'] == 1:
                playerObj['dribble']+=1
            elif event_id == 4 and rowObj['outcome'] == 0:
                playerObj['foulConceded']+=1
            elif event_id==7:
                playerObj['tackles']+=1
                if(rowObj['outcome'])==1:
                    playerObj['tackleSuccess'] += 1
            elif event_id==10:
                if(rowObj['def_block'])==1:
                    playerObj['blocks'] += 1
                else:
                    playerObj['saves'] += 1
            elif event_id == 17:
                if rowObj['yellow_card']==1:
                    playerObj['yellow_card']+=1
                if rowObj['second_yellow']==1 or rowObj['red_card']==1 :
                    playerObj['red_card']+=1
            playerDict[playerId] = playerObj      
    return playerDict
print('done')
```

```python
def createPenaltyAnalysisDataframe(sourceDf,playerDict):
    defObj = {'clubName':'','player_id':0, 'player':'', 'low_left':0,'high_left':0, 'low_centre':0, 
              'high_centre':0, 'low_right':0,'high_right':0,'close_left':0,'close_right':0,
              'close_high':0,'close_left_and_high':0,'saves':0,'close_right_and_high':0}
    data = sourceDf.groupby('game_id')
    for key, item in data:
        matchObj = data.get_group(key)
        matchObj = matchObj.dropna(subset=['player_id'])
        ctr=0
        for idx,rowObj in matchObj.iterrows():
            playerId = (rowObj['player_id'])
            if playerId == '':
                continue;
            else:
                playerId = int(playerId)
            if playerId in playerDict:
                playerObj = playerDict[(playerId)]
            else:
                playerObj = defObj.copy()

            if playerObj['player_id'] == 0:
                playerObj['player_id'] = playerId
            if playerObj['player'] == '':
                playerObj['player'] = rowObj['player']
            playerObj['clubName'] = rowObj['team']
            event_id = rowObj['event_type_id']
            if event_id in[13,16,15,14]:

                if rowObj['low_left'] == 1:
                    playerObj['low_left']+=1
                if rowObj['high_left'] == 1:
                    playerObj['high_left']+=1
                if rowObj['low_centre'] == 1:
                    playerObj['low_centre']+=1
                if rowObj['high_centre'] == 1:
                    playerObj['high_centre']+=1
                if rowObj['low_right'] == 1:
                    playerObj['low_right']+=1
                if rowObj['high_right'] == 1:
                    playerObj['high_right']+=1
                if rowObj['close_left'] == 1:
                    playerObj['close_left']+=1
                if rowObj['close_right'] == 1:
                    playerObj['close_right']+=1
                if rowObj['close_high'] == 1:
                    playerObj['close_high']+=1
                if rowObj['close_left_and_high'] == 1:
                    playerObj['close_left_and_high']+=1
                if rowObj['close_right_and_high'] == 1:
                    playerObj['close_right_and_high']+=1
                
            playerDict[playerId] = playerObj      
    return playerDict
print('done')
```
